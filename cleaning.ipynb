```
import csv
import os
from tika import parser
path = '.'
files = os.listdir(path)
for i in files:
  if "csv" in i:
    file_data = parser.from_file(i)
    text = file_data["content"]
    if text != None:
      text = text.encode('cp932', 'ignore')
      text = text.decode('cp932')
      f = open(i.replace("csv", "txt"), 'w')
      f.write(text)
      f.close()
```
